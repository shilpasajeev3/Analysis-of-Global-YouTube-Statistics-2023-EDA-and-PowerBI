#### Top 10 youtubers based on number of subscribers

```python
youtuber_sort=df.sort_values(by='subscribers',ascending=False)
top_youtubers=youtuber_sort.head(10)
subscribers=top_youtubers['subscribers']/1e6
result=pd.DataFrame({'Youtuber':top_youtubers['Youtuber'], 'Subscribers(Millions)':subscribers})
result
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(10, 6))
plt.xlabel('Youtubers')
plt.ylabel('Subscribers (Millions)')
plt.title('Top 10 youtubers based on number of subscribers',fontsize=20)
plt.xticks(rotation=90)
plt.bar(top_youtubers['Youtuber'], subscribers, color=['red','black'])
```

As we can observe, T-Series(India) has the highest subscriber count, with a massive 245 million subscribers. YouTube Movies (US), with 170 million subscribers, secures the second position, while Mr. Beast(US) follows closely with 166 million subscribers, ranking third.

#### Top 10 youtubers based on number of Video Views

```python
youtuber_sort=df.sort_values(by='video views',ascending=False)
top_youtubers=youtuber_sort.head(10)
video_views=top_youtubers['video views']/1e9
result=pd.DataFrame({'Youtuber':top_youtubers['Youtuber'], 'Video Views(Billions)':video_views})
result
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(10, 6))
plt.xlabel('Youtubers')
plt.ylabel('Video Views (Billions)')
plt.title('Top 10 youtubers based on number of Video Views',fontsize=20)
plt.xticks(rotation=90)
plt.bar(top_youtubers['Youtuber'], video_views, color=['red','black'])
```

T-Series takes the lead with an impressive 228 billion video views, securing the top position. Following closely is cocomelon - Nursery Rhymes with 164 billion views, claiming the second spot, while SET India takes the third position with 148 billion views.

#### Number of Youtubers based on Category

```python
category_count=df['category'].value_counts()
category_count
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(14, 6))
plt.xlabel('Category')
plt.ylabel('Youtubers')
plt.title('No.of youtubers based on category',fontsize=20)
plt.xticks(rotation=90)
plt.bar(category_count.index, category_count.values, color=['red','black'])
```

The majority of YouTubers are concentrated in categories such as Entertainment, Music, People & Blogs, and Gaming, while categories like Travel & Events, Nonprofits & Activism, Autos & Vehicles, Movies and Trailers have relatively fewer channels.

#### Number of Subscribers by Country

```python
country_subscriber=df.groupby('Country')['subscribers'].sum()
country_subscribers=country_subscriber.sort_values(ascending=False)
country_subscribers=country_subscribers/1e9
country_subscribers
country_subscribers.head()
```

```python
country_subscribers.tail()
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(14, 6))
plt.xlabel('Country')
plt.ylabel('Subscribers (Billions)')
plt.title('No.of Subscribers based on Country',fontsize=20)
plt.xticks(rotation=90)
plt.bar(country_subscribers.index, country_subscribers.values, color=['red','black'])
```

The highest number of subscribers comes from the United States, India, and Brazil, while Samoa, Finland, and Bangladesh have the fewest subscribers.

#### Video Views by Category


```python
category_view=df.groupby('category')['video views'].sum()
category_views=category_view.sort_values(ascending=False)
category_views=category_views/1e9
category_views
category_views.head()
```

```python
category_views.tail()
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(14, 6))
plt.xlabel('Category')
plt.ylabel('Video Views (Billions)')
plt.title('Video Views by Category',fontsize=20)
plt.xticks(rotation=90)
plt.bar(category_views.index, category_views.values, color=['red','black'])
```

Music, Entertainment, and People & Blogs categories have the highest number of views, while Travel & Events, Nonprofits & Activism, Autos & Vehicles, and Movies have comparatively fewer views.
